```python
# which seasons bring out the best human performance
```

```python
# seasons for the data

#spring:3-5
#summer:6-8
#fall:9-11
#winter:12-2
```

```python
# first create a month attribute
```

```python
ds3["race_month"]=ds3["race_day"].str.split(".").str.get(1).astype(int)
```

```python
ds3.head()
```

```python
#make a new season attribute
```

```python
ds3["race_season"]=ds3["race_month"].apply(lambda x: 'winter' if (x>11 or x<3) else "spring" if (x>2 and x<6) else "summer" if(x>5 and x<9) else "fall" )
```

```python
ds3.groupby("race_season")["athlete_average_speed"].mean().sort_values(ascending=False)
```

```python
#it shows that the summer season is the wort performing season whereas the spring season is the best performing
```

```python
# 4 showing seasonwise performance in the 50 mi race
```

```python
ds3.query('race_length=="50mi"').groupby("race_season")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean",ascending=False)
```

```python
# in 50 mi fall season is the best for performance
```
